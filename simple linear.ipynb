# Simple Linear Regression (house price prediction)

```python
#importing the libraries
import pandas as  pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#reading data(CSV format)
df=pd.read_csv("house_prices.csv")
```

```python
#checking first 5 rows
df.head()
```

```python
#plotting scatter plot to see relation between LotArea and SalePrice
plt.scatter(df.LotArea,df.SalePrice)
plt.xlabel('LotArea')
plt.ylabel("SalePrice")
plt.title('house pricing based on LotArea')
```

We can see some relation in above scatter plot. as the Lot area is increases the price of house will increases.

# Fitting the model

```python
#fitting simple regression to training set

reg=LinearRegression()
X=df[['LotArea']]
y=df.SalePrice
reg.fit(X,y)
```

```python
#Predicting the test set results
y_pr=reg.predict(X)
```

# Plotting best fitting line..

```python
plt.scatter(X,y)
plt.plot(X,y_pr,"r-")
plt.title('Housing Price ')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```

A line in Above plot showing best relation between Area and HousePrice.........

# Checking above plot and fitted model give apprx. same predictions or not...

```python
plt.scatter(X,y)
plt.plot(X,y_pr,"r-")
plt.title('Housing Price ')
plt.axvline(x=14000,color='r')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```
